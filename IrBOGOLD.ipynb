```python
import numpy as np

X = np.array([0, 1, 2, 3, 4, 5])
Y = np.array([0, 2.2, 3.9, 5.5, 7.9, 11])
```

```python
W = 0
b = 0
N = len(X)
```

```python
for i in range(10000):
    Y_pred = W * X + b

    dw = (2/N) * np.sum(X * (Y_pred - Y))
    db = (2/N) * np.sum(Y_pred - Y)

    W = W - 0.01 * dw
    b = b - 0.01 * db
```

```python
loss = np.mean(((W * X + b) - Y) ** 2)
loss
```
